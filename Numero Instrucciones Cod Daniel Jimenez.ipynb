```python
import random
```

```python
def randomPerm(n):
    v=[]
    #Construir un arrelo de 1 hasta n
    for i in range(n):
        v.append(i+1)
    # Revolver el arreglo con swaps aleatorios n veces
    for i in range(len(v)-1):
        j = random.randint(i, len(v)-1)
        aux = v[i]
        v[i] = v[j]
        v[j] = aux
    return v
```

```python
def isortSteps(a):
    #a es el arreglo con la permutacion aleatoria generada
    v = []
    #Construir un arreglo y le pasar toda la indormacion del arreglo "a" 
    for i in range(len(a)):
        v.append(a[i])
        
    steps = 0 #inicilizo contador de numero de pasos
    
    for i in range(1,len(v)): #paso 1 -for
        key = v[i] #paso 2 -for
        j = i-1 #paso3 -for
        while (j > -1) and (v[j] > key): #paso1 -while
            v[j+1] = v[j] #paso2 -while
            j = j -1 #paso3 -while
            steps = steps + 3
        steps = steps + 1   #== 1 paso cuando pregunta y no se cumple el while ==
        v[j+1] = key #paso 4 -for
        steps = steps + 4  # == 4 pasos del for ==
    steps = steps + 1  #== 1 paso cuando pregunta y no se cumple el for ==
    return steps
```

```python
n = 10  #tama√±o del arreglo que contendra la permutacion aleatoria.
runs = 100  #numero de permutaciones a las que se les calcularan los pasos o instrucciones.

sum = 0
min = n**3
max = 0

for i in range(runs):
    t = isortSteps(randomPerm(n)) 
    print "Numero instrucciones permutacion",i+1,":", t
    if t < min :
       min = t
    if t > max :
       max = t
    sum = sum + t
    
print 'Theoretical best time, ' + str(5*n - 4) 
print 'Theoretical worst time,' + str((3.0/2.0)*n**2 + (7.0/2.0)*n - 4)
print 'Theoretical average time,' + str((3.0/4.0)*n**2 + (17.0/4.0)*n - 4)
print 'Experimenal best time, ' + str(min)
print 'Experimenal worst time,' + str(max)
print 'Experimenal average time,' + str(sum/runs)
```
