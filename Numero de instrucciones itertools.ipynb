## Numero de instrucciones itertools

```python
import random
import itertools
```

```python
def allpermutations(n):
    v=[]
    #Construir un arrelo de 1 hasta n
    for i in range(n):
        v.append(i+1)
    
    p = list(itertools.permutations(v)) # lista con todas las permutaciones de 
    
    return p
```

```python
def isortCompare(a):
    #a es el arreglo con la permutacion aleatoria generada
    v = []
    
    #Construir un arreglo y le pasar toda la indormacion del arreglo "a" 
    for i in range(len(a)):
        v.append(a[i])
    
    #inicilizar contador de numero de comparaciones    
    nCompare = 0   
    for i in range(1,len(v)): 
        key = v[i] 
        j = i-1
        while (j > -1) and (v[j] > key): 
            v[j+1] = v[j] 
            j = j -1 
            nCompare = nCompare + 1
        nCompare = nCompare + 1
        v[j+1] = key 
        
    return nCompare
```

```python
n = 8  #tamao del arreglo que contendra la permutacion aleatoria.

sum = 0
min = n**3
max = 0

p = allpermutations(n) # Generar todas las permutaciones de n numeros

for i in range(len(p)):
    t = isortCompare(p[i]) 
    print "Numero instrucciones permutacion",i+1,":", t
    if t < min :
       min = t
    if t > max :
       max = t
    sum = sum + t
    
print 'Theoretical best time, ' + str(5*n - 4) 
print 'Theoretical worst time,' + str((3.0/2.0)*n**2 + (7.0/2.0)*n - 4)
print 'Theoretical average time,' + str((3.0/4.0)*n**2 + (17.0/4.0)*n - 4)
print 'Experimenal best time, ' + str(min)
print 'Experimenal worst time,' + str(max)
print 'Experimenal average time,' + str(sum/len(p))
```
