```python
import random
```

```python
def randomPerm(n):
    v=[]
    #Construir un arrelo de 1 hasta n
    for i in range(n):
        v.append(i+1)
    # Revolver el arreglo con swaps aleatorios n veces
    for i in range(len(v)-1):
        j = random.randint(i, len(v)-1)
        aux = v[i]
        v[i] = v[j]
        v[j] = aux
    return v
```

```python
def isortSwap(a):
    #a es el arreglo con la permutacion aleatoria generada
    v = []
    
    #Construir un arreglo y le pasar toda la indormacion del arreglo "a" 
    for i in range(len(a)):
        v.append(a[i])
    
    #inicilizar contador de numero de comparaciones    
    nSwap = 0   
    for i in range(1,len(v)): 
        key = v[i] 
        j = i-1
        while (j > -1) and (v[j] > key): 
            v[j+1] = v[j] 
            j = j -1 
            nSwap = nSwap + 1

        v[j+1] = key 
        
    return nSwap
```

```python
n = 50  #tamao del arreglo que contendra la permutacion aleatoria.
runs = 5000  #numero de permutaciones a las que se les calcularan los pasos o instrucciones.

sum = 0
min = n**3
max = 0

for i in range(runs):
    t = isortSwap(randomPerm(n)) 
    print "Numero swaps permutacion",i+1,":", t
```
